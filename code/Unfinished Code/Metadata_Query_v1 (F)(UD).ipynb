```python
def origin_dir():

    home_dir_list = os.listdir( '/home' )

    for dir in home_dir_list:

        if '_linux' in dir:

            path = f'/home/{dir}/Desktop/TDMprivate'

            if not os.path.exists( path ):

                raise Exception( 'TDMprivate folder does not exist. TDMprivate folder must exist on Desktop. Notify Ryan or ' +
                                 'Vincent if this message appears.' )

            else:

                return path
```

```python
if os.path.exists( f'{ origin_dir() }/metadata' ):

    has_folder = True

    if os.path.isfile( f'{ origin_dir() }/metadata/metadata.csv' ):

        answer = None

        while ( ( answer != 'Y' ) and ( answer != 'N' ) ):

            answer = input( 'metadata.csv file already exists. Overwrite? (Y/N)' )

        if ( answer == 'N' ):

            raise Exception( 'Program aborted.' )

else:

    has_folder = False

#

dynamodb = boto3.resource( 'dynamodb' )

adsppm_table = dynamodb.Table( 'ads_passenger_processed_metadata' )

gmID_metadata_dfs_list = []

for gmID in gmID_list:

    query_input = dict(
                       IndexName = 'groupMetadataID-index',

                       KeyConditionExpression = Key( 'groupMetadataID' ).eq( gmID ),
                       
                       ProjectionExpression = 'vehicleID, Route, #operator, driver, groupMetadataID, experimentID, #other',

                       ExpressionAttributeNames = {
                                                   '#operator' : 'operator',
                      
                                                   '#other' : 'other'
                                                  }
                      )

    start = True

    query_output = {}

    while ( start or ( 'LastEvaluatedKey' in query_output.keys() ) ):

        if not start:

            query_input[ 'ExclusiveStartKey' ] = query_output[ 'LastEvaluatedKey' ]

        start = False

        query_output = adsppm_table.query( **query_input )

        if query_output[ 'Items' ]:

            query_output_df = pd.DataFrame.from_dict( pd.json_normalize( query_output[ 'Items' ] ), orient = 'columns' )

            gmID_metadata_dfs_list = gmID_metadata_dfs_list + [ query_output_df ]

            break

    print( f'{len( gmID_metadata_dfs_list )} / 330 gmIDs Queried' )

    clear_output( wait = True )

#

metadata_df = pd.concat( gmID_metadata_dfs_list )

metadata_df = metadata_df.drop_duplicates()

if not has_folder:

    os.mkdir( f'{ origin_dir() }/metadata' )

metadata_df.to_csv( f'{ origin_dir() }/metadata/metadata.csv', index = False )
```
