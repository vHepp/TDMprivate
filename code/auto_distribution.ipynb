```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ads_index = pd.read_csv("./data/ads_data_index_w_percentage.csv")

ads_index.head()
```

```python
temp =  ads_index.copy()
temp['datetime'] = pd.to_datetime(temp['end_time'])

temp.sort_values('datetime', inplace=True)
temp = temp[:]
#temp = temp

fig, ax = plt.subplots(figsize=(12, 3), layout='constrained')

ax.plot(temp['datetime'],[1]*temp.shape[0],color="black", linewidth=.2, linestyle='--')
ax.plot(temp['datetime'],[0]*temp.shape[0],color="black", linewidth=.2, linestyle='--')
ax.plot(temp['datetime'],temp["percent_auto"], marker='o', ms=5,mec='blue', mfc="blue", color="orange")

ax.set_ylabel("% of run autonomous")
ax.set_xlabel("run end_time")

plt.show()
```
