```python
import CalculatedFieldSubroutines as cfs
import pandas as pd
import numpy as np
import os
import plotly.express as px
```

### Read and preprocces data


```python
ads_index = pd.read_csv('./data/ads_data_index.csv')
ads_index
```

```python
df_chassis = pd.DataFrame()

for gmID in ads_index["groupMetadataID"]:
    print(gmID)
    temp = pd.read_csv(f'./data/Raw_Data/{gmID}/_apollo_canbus_chassis/{gmID}_apollo_canbus_chassis.csv')
    
    cfs.BinaryDrivingMode(temp)
    cfs.TernaryDrivingModeTransition(temp)
    
    subset = temp[["groupMetadataID","time", "BinaryDrivingMode","TernaryDrivingModeTransition"]].copy()
    
    df_chassis = pd.concat([df_chassis,subset])
```

```python
d = []

for gmID in ads_index["groupMetadataID"]:
    print(gmID)
    
    o = df_chassis[df_chassis["groupMetadataID"] == gmID]["BinaryDrivingMode"].value_counts().sort_index()
    d.append({"groupMetadataID": gmID, "object":o})
    
    
    # print({"id": gmID, "unique":})
    #print(o)
    
```

```python
p = pd.DataFrame(d)
p
```

```python
#p["percent_manual"] = p['unique'][1]/(p['unique'][1]+p['unique'][])

auto_nums = []
manual_nums = []
modes = []
percents = []
isOneMode = []


for index,row in p.iterrows():
    mode_list = list(row.iloc[1].index)
    #print(mode_list)
    if len(mode_list) == 1:
        
        isOneMode.append(True)
        
        if 0 in mode_list:
            percents.append(0.0)
            modes.append(0)
            auto_nums.append(0)
            manual_nums.append(row.iloc[1].iloc[0])
            print("only manual", index)
        elif 1 in mode_list:
            percents.append(1.0)
            modes.append(1)
            manual_nums.append(0)
            auto_nums.append(row.iloc[1].iloc[0])
            print("only auto")
    elif len(mode_list) > 1:
        isOneMode.append(False)  
        percent_auto =  row.iloc[1].iloc[1] / (row.iloc[1].iloc[0] + row.iloc[1].iloc[1])
        
        percents.append(percent_auto)
        modes.append("both")
        manual_nums.append(row.iloc[1].iloc[0])
        auto_nums.append(row.iloc[1].iloc[1])
        
p['isOneMode'] = isOneMode
p["percent_auto"] = percents
p['modes'] = modes
p["auto_count"] = auto_nums
p["manual_count"] = manual_nums
```

```python
final = p[["groupMetadataID","isOneMode","modes","auto_count","manual_count", "percent_auto"]].copy()

final
```

```python
final_merged = pd.merge(ads_index, final, on="groupMetadataID")

final_merged
```

```python
final_merged.to_csv("./data/ads_data_index_w_percentage.csv", index=False)
```

## Calculate total observations

```python
ads_index = pd.read_csv('./data/ads_data_index_w_percentage.csv')
ads_index.head()
```

```python
ads_index["total_count"] = ads_index['auto_count'] + ads_index["manual_count"]

cols = ['groupMetadataID',
 'start_time',
 'end_time',
 #'nummsgs',
 #'filebase',
 #'folder',
 'vehicleID',
 #'experimentID',
 'route',
 'isOneMode',
 #'modes',
 'auto_count',
 'manual_count',
 'percent_auto',
 'total_count'
 ]
```

```python
ads_index[cols]
```

```python
avgs = pd.DataFrame()
avgs["avg"] = [-1] * ads_index.shape[0]
avgs["route"] = [""] * ads_index.shape[0]
avgs["avg"]
```

```python
red_indexs = ads_index[(ads_index["route"] == "RedRoute")].index
red_indexs

reds = ads_index.iloc[red_indexs][cols].sort_values("total_count")
red_mean = reds["total_count"].describe()["mean"]
red_mean


avgs["route"].iloc[red_indexs] = "Red"
avgs["avg"].iloc[red_indexs] = red_mean
avgs.head(20)
```

```python
blue_indexs = ads_index[(ads_index["route"] == "Blue Route")].index
blue_indexs

blues = ads_index.iloc[blue_indexs][cols].sort_values("total_count")
blue_mean = blues["total_count"].describe()["mean"]
blue_mean

avgs["route"].iloc[blue_indexs] = "Blue"
avgs["avg"].iloc[blue_indexs] = blue_mean
avgs.head(20)
```

```python
green_indexs = ads_index[(ads_index["route"] == "GreenRoute")].index
green_indexs

greens = ads_index.iloc[green_indexs][cols].sort_values("total_count")
green_mean = greens["total_count"].describe()["mean"]
green_mean

avgs["route"].iloc[green_indexs] = "Green"
avgs["avg"].iloc[green_indexs] = green_mean
avgs.head(20)
```

```python
ads_index["route avg total_count"] = avgs["avg"]

ads_index["isAboveAvg"] = ads_index["total_count"] > ads_index["route avg total_count"]

ads_index.head(20)
```

```python
ads_index.to_csv("./data/ads_data_index_w_percentage.csv", index=False)
```
